```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.ensemble import RandomForestRegressor
import joblib

# 1. Load our data
df = pd.read_csv('data/student_expenses.csv')

# 2. THE HIGH-LEVEL LOGIC: The "Prescriptive" Advice Engine
def generate_advice(row):
    if row['online_food_freq'] > 12:
        return "Critical: Online food frequency is too high. Switching to home-cooked meals could save 30% of your budget."
    elif row['exam_month'] == 1 and row['food_exp'] > 200:
        return "Exam Alert: Stress spending detected. Try meal prepping to avoid expensive last-minute orders."
    elif row['prev_month_expense'] > row['monthly_allowance']:
        return "Budget Overrun: You are spending more than your allowance. Emergency budget re-routing suggested."
    else:
        return "Healthy: Your spending is on track. Suggested: Move 10% to savings."

# Apply the AI logic
df['ai_advice'] = df.apply(generate_advice, axis=1)

# 3. VISUALIZATION (This makes it look high-level)
fig = px.scatter(df, x="monthly_allowance", y="total_expense", 
                 color="ai_advice", title="AI Expense Analysis Dashboard",
                 hover_data=['student_type', 'online_food_freq'])
fig.show()

print("✅ Analysis Complete. Visual Dashboard Generated.")
```

```python
import pandas as pd
import numpy as np
import os

# Create the directory if it doesn't exist
if not os.path.exists('data'):
    os.makedirs('data')

# Generate the High-Level Synthetic Dataset
np.random.seed(42)
data = {
    'student_id': range(1, 1001),
    'student_type': np.random.choice(['Hosteller', 'Day Scholar'], 1000),
    'monthly_allowance': np.random.randint(300, 1200, 1000),
    'prev_month_expense': np.random.randint(200, 1100, 1000),
    'food_exp': np.random.randint(50, 400, 1000),
    'shopping_exp': np.random.randint(20, 250, 1000),
    'transport_exp': np.random.randint(10, 150, 1000),
    'online_food_freq': np.random.randint(0, 25, 1000),
    'exam_month': np.random.choice([0, 1], 1000, p=[0.8, 0.2]),
}

df_init = pd.DataFrame(data)
# Logic: Total expense is a mix of categories + a 'stress' penalty for exams
df_init['total_expense'] = (df_init['food_exp'] + df_init['shopping_exp'] + 
                            df_init['transport_exp'] + (df_init['exam_month'] * 75))

# Save it so the next cell can find it
df_init.to_csv('data/student_expenses.csv', index=False)

print("✅ Data File Created Successfully at 'data/student_expenses.csv'!")
```
